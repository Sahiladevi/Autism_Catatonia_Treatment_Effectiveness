# Data visualization for Autism and catatonia treatment effectiveness

## 1. Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#from io import StringIO
```

## 2. Load data

```python
# Create the dataframe
script_dir = os.getcwd()  # get current working directory

file_path = os.path.join(script_dir, './data/autism_catatonia_medication_simulated.xlsx')

df = pd.read_excel(file_path)

# To show all rows regardless of count
pd.set_option('display.max_rows', None)
df
```

## 3. Data Visualization

### A. Improvement (%) Comparison

 1. Bar chart  - Improvement (%) by ID after treatment - Improvement (%) Comparison

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Adjusted figure size
plt.bar(df['ID'], df['Improvement (%)'], color='deepskyblue', edgecolor='black', width=0.5)

# Add labels and title
plt.xlabel('ID')
plt.ylabel('Improvement (%)')
plt.title('Improvement (%) by ID after Treatment')

# Rotate x-axis labels for clarity
plt.xticks(rotation=45)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

2. Box Plot of Improvement (%) grouped by Medication.

```python
# Plotting the box plot grouped by Medication
plt.figure(figsize=(8, 6))
sns.boxplot(x='Medication', y='Improvement (%)', data=df, palette="Set2", hue='Medication', legend=False)

# Add labels and title
plt.ylabel('Improvement (%)')
plt.xlabel('Medication')
plt.title('Improvement (%) Grouped by Medication')

# Display the plot
plt.tight_layout()
plt.show()
```

3. Box plot of Improvement (%) grouped by  Catatonia_Diagnosis.

```python
# Plotting the box plot grouped by Catatonia_Diagnosis
plt.figure(figsize=(8, 6))
sns.boxplot(x='Catatonia_Diagnosis', y='Improvement (%)', data=df, palette="Set2", hue='Catatonia_Diagnosis', legend=False)

# Add labels and title
plt.xlabel('Catatonia Diagnosis')
plt.ylabel('Improvement (%)')
plt.title('Improvement (%) Grouped by Catatonia Diagnosis')

# Display the plot
plt.tight_layout()
plt.show()
```

### B. Medication Efficacy

4. A Box Plot or Violin Plot of Improvement (%) grouped by the Medication type.

```python
# Set style
sns.set(style="whitegrid")

# ----------- Box Plot and Violin Plot by Medication -----------

# Optional: Exclude 'None' meds if focusing on active treatments
active_meds_df = df[df['Medication'] != 'None']

plt.figure(figsize=(14, 6))

# Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Medication', y='Improvement (%)',  hue='Medication', data=active_meds_df, palette='Set3')
plt.title('Box Plot: Improvement by Medication')
plt.xticks(rotation=45)

# Violin Plot
plt.subplot(1, 2, 2)
sns.violinplot(x='Medication', y='Improvement (%)', hue='Medication', data=active_meds_df, palette='Set3')
plt.title('Violin Plot: Improvement by Medication')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

5. Scatter Plot: Dosage (mg) vs Improvement (%) - Showing Medication Efficacy

```python
# Plotting the scatter plot of Dosage_mg vs Improvement (%)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Dosage_mg', y='Improvement (%)', data=df, hue='Medication', palette="Set2", s=100)

# Add labels and title
plt.xlabel('Dosage (mg)')
plt.ylabel('Improvement (%)')
plt.title('Scatter Plot: Dosage (mg) vs Improvement (%)')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import pearsonr

# Calculate Pearson correlation
correlation, p_value = pearsonr(df['Dosage_mg'], df['Improvement (%)'])

# Print correlation result
print(f"Pearson correlation coefficient: {correlation:.2f}")
print(f"P-value: {p_value:.4f}")

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Dosage_mg', y='Improvement (%)', data=df, scatter_kws={'s':100}, line_kws={'color':'red'}, ci=None)

# Annotate correlation on the plot
plt.text(0.5, max(df['Improvement (%)']) - 5, f'Pearson r = {correlation:.2f}', fontsize=12, color='blue')

# Labels and title
plt.xlabel('Dosage (mg)')
plt.ylabel('Improvement (%)')
plt.title('Dosage vs Improvement with Regression Line')

plt.tight_layout()
plt.show()
```

## C. Catatonia Diagnosis Impact

6. Box Plot or Violin Plot:
This compares the distribution of Improvement (%) between two groups — those with and without a Catatonia_Diagnosis.

```python

# Set Seaborn style
sns.set(style="whitegrid")

# --------- 1. Box Plot or Violin Plot ---------
plt.figure(figsize=(12, 5))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Catatonia_Diagnosis', y='Improvement (%)',hue='Catatonia_Diagnosis', data=df, palette='Set2')
plt.title("Box Plot: Improvement by Catatonia Diagnosis")

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(x='Catatonia_Diagnosis', y='Improvement (%)',hue='Catatonia_Diagnosis', data=df, palette='Set2')
plt.title("Violin Plot: Improvement by Catatonia Diagnosis")

plt.tight_layout()
plt.show()
```

7. Grouped Bar Chart:
This shows the average Improvement (%) for each group (Catatonia_Diagnosis = Yes and No).

```python
# Calculate average improvement
avg_improvement = df.groupby('Catatonia_Diagnosis')['Improvement (%)'].mean().reset_index()

# Bar chart
plt.figure(figsize=(6,5))
sns.barplot(x='Catatonia_Diagnosis', y='Improvement (%)',hue ='Catatonia_Diagnosis', data=avg_improvement, palette='pastel')
plt.title("Average Improvement by Catatonia Diagnosis")
plt.ylabel("Average Improvement (%)")
plt.xlabel("Catatonia Diagnosis")
plt.ylim(0, 90)  # optional, adjust y-axis for better visuals
plt.show()
```

### D. 4. Side Effects Analysis

8. Bar Chart of Side_Effects counts to see which are most common

```python
# Count the occurrences of each side effect
side_effect_counts = df['Side_Effects'].value_counts()

# Plot the bar chart
plt.figure(figsize=(12, 6))
side_effect_counts.plot(kind='bar', color='skyblue')
plt.title('Most Common Side Effects')
plt.xlabel('Side Effect')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

9. Box Plot: Improvement (%) grouped by presence or absence of side effects

```python

# Create a new column for presence of side effects
df['Has_Side_Effects'] = df['Side_Effects'].notna() & (df['Side_Effects'].str.strip() != '')

# Optional: Convert to Yes/No for clarity in plot
df['Side Effects Present'] = df['Has_Side_Effects'].map({True: 'Yes', False: 'No'})

# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Side Effects Present', y='Improvement (%)',hue='Side Effects Present', palette='pastel')

plt.title('Improvement (%) by Side Effects Presence')
plt.xlabel('Side Effects Present')
plt.ylabel('Improvement (%)')
plt.tight_layout()
plt.show()
```

### E. 5. Gender & Age Analysis

### E. Treatment Duration

12. Scatter Plot: Duration_Days vs Improvement (%) to explore if longer treatment leads to better improvement.

```python

# Filter out rows with Duration = 0 (i.e., no treatment)
treated_df = df[df['Duration_Days'] > 0]

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=treated_df,
    x='Duration_Days',
    y='Improvement (%)',
    hue='Medication',  # Optional: color by medication type
    palette='Set2',
    s=100  # Marker size
)

# Optional: Add a regression trend line
sns.regplot(
    data=treated_df,
    x='Duration_Days',
    y='Improvement (%)',
    scatter=False,
    color='black',
    line_kws={"linestyle":"--"}
)

plt.title('Duration of Treatment vs. Improvement (%)')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Improvement (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

13. Plot Pre_Treatment_Score and Post_Treatment_Score.

```python
#group by medication
df_top = df.groupby('Medication').apply(
    lambda group: group.sort_values('Improvement (%)', ascending=False,).head(5)
).reset_index(drop=True)

# Melt the data
df_melted = df_top.melt(
    id_vars=['ID', 'Medication'],
    value_vars=['Pre_Treatment_Score', 'Post_Treatment_Score'],
    var_name='Treatment_Stage',
    value_name='Score'
)

# Plot using facet grid
g = sns.catplot(
    data=df_melted,
    x='ID', y='Score', hue='Treatment_Stage',
    col='Medication', kind='bar',
    col_wrap=3, height=4, aspect=1.5,
    palette='muted'
)

g.set_titles("Medication: {col_name}")
g.set_xticklabels(rotation=45)
plt.tight_layout()
plt.show()
```
